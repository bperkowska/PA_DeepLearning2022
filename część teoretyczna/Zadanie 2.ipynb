```python
import random
import numpy as np
random.seed(1)
```

```python
A = np.matrix([random.random() for i in range(300)]).reshape(30,10) # ma 30 wierszy, 10 kolumn (każda lista w liście to jeden wiersz)
At = np.transpose(A) # 10 wierszy, 30 kolumn
b = np.matrix([random.random() for i in range(30)]).reshape(30,1)  # 30 wierszy, 1 kolumna
```

```python
b.view(np.float64)
```

```python
A.view(np.float64)
```

```python
At.view(np.float64)
```

```python
# calculating least squares approximate solution

x_hat = (At * A).I * At * b
x_hat  # 10 rows, 1 column
```

```python
# calculating the residual norm squared

norm_least_squared = np.square(A *x_hat - b)
least_sum = np.sum(norm_least_squared)
least_sum
```

```python
# generating random three vectors

d1 = np.matrix([random.random() for i in range(10)]).reshape(10,1)
d2 = np.matrix([random.random() for i in range(10)]).reshape(10,1)
d3 = np.matrix([random.random() for i in range(10)]).reshape(10,1)
```

```python
print(d1)
```

```python
print(d2)
```

```python
print(d3)
```

```python

#checking whether stated condition holds

for i in [d1, d2, d3]:
    
    y = np.square(A * (x_hat + i) - b)
    y_sum = np.sum(y)
    if y_sum > least_sum:
        print(f"{y_sum} > {least_sum}")
        print("True")
```
